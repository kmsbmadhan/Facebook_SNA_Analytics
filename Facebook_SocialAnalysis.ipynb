```python
import networkx as nx
import matplotlib.pyplot as plt
import os
%matplotlib inline
os.listdir()
```

```python
#Read in file
fb = nx.read_edgelist('facebook_combined.txt',create_using=nx.Graph(),nodetype=int)
```

```python
print(nx.info(fb))
```

```python
#Create a spring layout
pos = nx.spring_layout(fb)

plt.figure(figsize=(10,7))

plt.axis('off')

nx.draw_networkx(fb,node_size=4039,node_color='blue',with_labels=True,pos=pos)
```

```python
print(nx.betweenness_centrality(fb,normalized=True))
bc = nx.betweenness_centrality(fb)
plt.bar(range(len(bc)),bc.values(),align='center')
```

```python
print(fb.degree())
```

```python
print(fb.neighbors())
```

```python
print(fb.order())
print(fb.size())
```

```python
print(nx.triangles(fb))
```

```python
print(nx.clustering(fb))
```

```python
hst = nx.degree_histogram(fb)
plt.figure(figsize=(10,10))
plt.hist(hst,bins=10,color='c')
```
